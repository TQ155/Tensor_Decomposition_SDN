```python
# create excel files  
dpath = r'C:\Users\TQ\Desktop\Thesis_Code\weekly_data'
path_gb = r'C:\Users\TQ\Desktop\Thesis_Code\weekly_data_gb'


# this code create excel files from , to , amount and store it in weekly_data_gb folder 

from os import listdir, path
import os as os 
import pandas as pd 
count = 0 

def clean_s(df):
    
    df.columns = df.columns.str.replace(' ', '')
    df.columns = df.columns.str.lstrip()
    df.columns = df.columns.str.rstrip()
    df.columns = df.columns.str.strip()
    
    return df

    
for index , filename in enumerate(listdir(dpath)):
    df = pd.read_csv( dpath + "\\" + filename )
    df.drop(columns=df.columns[0], axis=1, inplace=True)
    df.columns = ["from" , "to" , "amount"]
    df['amount'] = df['amount'].astype(float)
    clean_s(df)
    print("Week #" + str(index + 1) )
    df = df.groupby(["from","to"])["amount"].sum().reset_index()
    df.to_excel(os.path.join(path_gb, "week{}".format(str(index + 1) ) + ".xlsx"), index = False)
    
    print(df)
    
```
